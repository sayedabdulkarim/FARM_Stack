```python
short_data = data[["id","name","brand", "categories", "dateAdded", "prices.amountMax", "manufacturerNumber", "colors", "sourceURLs"]]
```

```python
short_data
```

```python
short_data
```

```python
import numpy as np
 
# print the list of 10 integers from 3  to 7
stocks = list(np.random.randint(low = 20,high=100,size=short_data.shape[0]))
 
# stocks = random.sample(range(20, 100), short_data.shape[0])
short_data["stocks"] = stocks
```

```python
short_data["colors"] = short_data["colors"].fillna("Red")
```

```python
short_data["brand"] = short_data["brand"].fillna("PUMA")
short_data["manufacturerNumber"] = short_data["manufacturerNumber"].fillna("ZSR101BLMLG")
```

```python
short_data.rename(columns = {'categories':'Description', 'dateAdded':'date of manufacture', 'prices.amountMax':'price'}, inplace = True)
```

```python
short_data
```

```python
short_data.isna().any().any()
```

```python
product_catalogue_dict = short_data.to_dict('records')
```

```python
product_catalogue_dict
```

```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')

with client:

    db = client.products_catalogue
    
    db.products_info.insert_many(product_catalogue_dict)
    # db = client.product_catalogue

    
```
